```python
import matplotlib.pyplot as plt
import numpy as np

# Data
years = np.arange(2010, 2020)
kohli_runs = [1200, 1380, 1420, 1500, 1320, 1550, 1650, 1700, 1450, 1600]
rohit_runs = [800, 950, 1100, 1200, 1300, 1450, 1480, 1500, 1400, 1550]
sehwag_runs = [1100, 1150, 1050, 980, 970, 800, 650, 400, 200, 100]

# Use a built-in custom style
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(12, 6))

# Plotting
plt.plot(years, kohli_runs, label='Virat Kohli', color='red', linestyle='-', linewidth=2)
plt.plot(years, rohit_runs, label='Rohit Sharma', color='blue', linestyle='--', linewidth=2)
plt.plot(years, sehwag_runs, label='Virender Sehwag', color='green', linestyle='-.', linewidth=2)

# Labels and title
plt.title('Comparison of Runs Scored (20102019)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Runs Scored', fontsize=12)

# Add legend
plt.legend()

# Display grid and layout
plt.grid(True)
plt.tight_layout()

# Show plot
plt.show()
```
