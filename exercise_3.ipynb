```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data: runs per selected year
years = ['2001', '2003', '2005', '2007', '2010']
sachin_runs = [1600, 1200, 900, 1100, 1400]
kohli_runs = [0, 0, 500, 1000, 1300]  # Kohli debuted later
sehwag_runs = [1100, 1300, 1000, 900, 1200]

# Side-by-side bar chart
x = np.arange(len(years))
width = 0.25

plt.figure(figsize=(10, 6))
plt.bar(x - width, sachin_runs, width, label='Sachin')
plt.bar(x, kohli_runs, width, label='Kohli')
plt.bar(x + width, sehwag_runs, width, label='Sehwag')

plt.xlabel('Year')
plt.ylabel('Runs')
plt.title('Runs Comparison (Selected Years)')
plt.xticks(x, years)
plt.legend()
plt.tight_layout()
plt.show()

# Horizontal bar chart: Total runs in debut 5 years
debut_runs = {
    'Sachin': sum([500, 700, 1200, 800, 1000]),  # 1989-1993
    'Kohli': sum([200, 800, 1000, 1200, 900]),   # 2008-2012
    'Sehwag': sum([400, 1000, 1100, 900, 850])   # 1999-2003
}

players = list(debut_runs.keys())
runs = list(debut_runs.values())

plt.figure(figsize=(8, 4))
plt.barh(players, runs, color=['blue', 'green', 'orange'])
plt.xlabel('Total Runs')
plt.title('Total Runs in Debut 5 Years')
plt.tight_layout()
plt.show()
```
