```python
import numpy as np

# Task 1: Create a 3×3 array filled with random numbers and print its shape
array_random = np.random.rand(3, 3)
print("3x3 Random Array:\n", array_random)
print("Shape:", array_random.shape)

# Task 2: Convert an array of floats [1.1, 2.2, 3.3] into integers
float_array = np.array([1.1, 2.2, 3.3])
int_array = float_array.astype(int)
print("Converted to Integers:", int_array)

# Task 3: Use fancy indexing to extract even numbers from [1, 2, 3, 4, 5, 6]
original_array = np.array([1, 2, 3, 4, 5, 6])
even_indices = [i for i, x in enumerate(original_array) if x % 2 == 0]
even_numbers = original_array[even_indices]
print("Even Numbers (Fancy Indexing):", even_numbers)

# Task 4: Reshape a 1D array of size 9 into a 3×3 matrix
array_1d = np.arange(9)
array_reshaped = array_1d.reshape(3, 3)
print("Reshaped to 3x3:\n", array_reshaped)

# Task 5: Use boolean masking to filter numbers greater than 50 in an array
sample_array = np.array([10, 55, 32, 77, 49, 100])
filtered_array = sample_array[sample_array > 50]
print("Numbers Greater Than 50:", filtered_array)
```
